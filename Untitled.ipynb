```python
from collections import defaultdict
import cv2
import numpy as np
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO('best.pt')

# Open the video file
video_path = "track-example.mkv"
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
frame_rate = int(cap.get(5))

# Define the codec and create a VideoWriter object
output_path = "tracked2.avi"  # You can change the file format if needed
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Use XVID codec
out = cv2.VideoWriter(output_path, fourcc, frame_rate, (frame_width, frame_height))

from collections import defaultdict
import cv2
import numpy as np
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO('best.pt')

# Open the video file
video_path = "track-example.mkv"
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
frame_rate = int(cap.get(5))

# Define the codec and create a VideoWriter object
output_path = "output_video.mp4"  # You can change the file format if needed
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Use XVID codec
out = cv2.VideoWriter(output_path, fourcc, frame_rate, (frame_width, frame_height))

# Store the track history
track_history = defaultdict(lambda: [])

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 tracking on the frame, persisting tracks between frames
        results = model.track(frame, persist=True)

        # Get the boxes and track IDs
        boxes = results[0].boxes.xywh.cpu()
        if len(boxes) > 0 and (boxes is not None) and (results[0].boxes.id is not None):
            print(boxes)


            track_ids = results[0].boxes.id.int().cpu().tolist()

        # Visualize the results on the frame
            annotated_frame = results[0].plot()
        # Check if any objects were detected
            # Plot the tracks
            for box, track_id in zip(boxes, track_ids):
                x, y, w, h = box
                track = track_history[track_id]
                track.append((float(x), float(y)))  # x, y center point
                if len(track) > 30:  # retain 90 tracks for 90 frames
                    track.pop(0)

                # Draw the tracking lines
                points = np.hstack(track).astype(np.int32).reshape((-1, 1, 2))
                cv2.polylines(annotated_frame, [points], isClosed=False, color=(0, 255, 0), thickness=10)

        # Write the annotated frame to the output video
        out.write(annotated_frame)

        # Display the annotated frame (optional)
        #cv2.imshow("YOLOv8 Tracking", annotated_frame)

    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()

# Release the VideoWriter
out.release()


 
```

```python
import os
from tqdm.auto import tqdm
answer_list = []
# Set the directory path where your image files are located
directory_path = "./test_dataset/dataset/"

# List of valid image file extensions (you can customize this list)
valid_extensions = (".jpg", ".jpeg", ".png", ".gif", ".bmp")
file_list = sorted(os.listdir(directory_path))
print(file_list)
# Iterate over the files in the directory
for filename in tqdm(file_list):
    # Check if the file has a valid image file extension
    if filename.lower().endswith(valid_extensions):
        # You can perform actions on the image file here, such as processing it or printing its name.
        print("Image file found: " + filename)
        img = cv2.imread(os.path.join(directory_path,filename))
        result = model.predict(img, save=True,imgsz=640, conf=0.4)
        if len(result)==0 or (result==None) or (result[0].boxes.xyxy.cpu() is None):
            answer_list.append(0)
        else:
            boxes_xyxy = result[0].boxes.xyxy.cpu().numpy()[0]
            new_box = img[int(boxes_xyxy[1]):int(boxes_xyxy[3]),int(boxes_xyxy[0]):int(boxes_xyxy[2])]
            ocr_result = reader.recognize(new_box)
            if ocr_result is not None:
                answer_list.append(str(ocr_result[0][-2]))
                print(answer_list[-1])
            else:
                answer_list.append(0)

        
```
